```python
#import dependecies
import pandas as pd
import matplotlib.pyplot as plt 

#reading Csv
readWheatIndexCSV = "Resources/S&PGSCIGrainsIndex12-07-2022.csv"
#turning into DF
wheatCashPricesDF = pd.read_csv(readWheatIndexCSV)
wheatCashPricesDF
```

```python
#transforming DF
wheatCashPricesDF = wheatCashPricesDF.rename(columns={"Symbol: $SG5F":"Date",
                                                        "Unnamed: 1":"Opening Price",
                                                        "Unnamed: 2":"High of the Day",
                                                        "Unnamed: 3": "Low of the Day",
                                                        "Unnamed: 4": "Close of the Day",
                                                        "Unnamed: 5": "Change of the Day"
                                                        })



wheatCashPricesDF.head()
```

```python
#creating new df with selected columns, then sorting the new df by date ascending
transformedCashPricesDF = wheatCashPricesDF[["Date", "Opening Price", "High of the Day", "Low of the Day", "Close of the Day", "Change of the Day"]]
transformedCashPricesDF = transformedCashPricesDF.sort_values(by="Date")
transformedCashPricesDF
```

```python
""" marketCatastopheDF = transformedCashPricesDF.loc[(transformedCashPricesDF["Date"] >= '2022-02-24'
                                                    & transformedCashPricesDF["Date"] <= '2022-03-10')] """
#converted the date columnto date time64 then queried a specific month into a new df 
transformedCashPricesDF["Date"] = transformedCashPricesDF["Date"].astype('datetime64[ns]')
marketCatastopheDF = transformedCashPricesDF.query("Date >= '2022-02-10' \
                       and Date < '2022-03-10'")
marketCatastopheDF.head(30)
```

```python
marketCatastopheDF2 = marketCatastopheDF[['Date', "Opening Price", "Close of the Day","Change of the Day"]]

marketCatastopheDF2.head()
```

```python
marketCatastopheDF2.dtypes
```

```python
#creating graphs from already existing DFs
x = marketCatastopheDF2['Date']
y = marketCatastopheDF2['Change of the Day']
fig1, ax1 = plt.subplots(figsize=(10, 5))
plt.title('Price Action for the $SCGI Grain Index during the dates of 2-10-2022 through 3-10-2022')
plt.xlabel("Date")
plt.ylabel("Daily Change in Price")
plt.plot(x,y)
""" rus_attacks = plt.axvline(x="2022-02-25", color='r', label='Russia Attacks Ukraine')
plt.annotate("Russia Attacks Ukraine-->",("2022-02-22"),fontsize=18,color="red") """
plt.savefig("Graphs/catastrophemonth.png")
plt.show()
```

```python
beforeTheWarDF = marketCatastopheDF2.query("Date >= '2022-02-10' \
                       and Date <= '2022-02-24'")
beforeTheWarDF
```

```python
x = beforeTheWarDF['Date']
y = beforeTheWarDF['Change of the Day']
fig1, ax1 = plt.subplots(figsize=(10, 5))
plt.title('Price Action for the $SCGI Grain Index before the Ukraine war start')
plt.xlabel("Date")
plt.ylabel("Daily Change in Price")
plt.plot(x,y)
plt.savefig("Graphs/catastrophebefore.png")
plt.show()
```
